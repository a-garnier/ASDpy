<h1>Generate folders + Convert wav to spectrograms for the cnn v2 model</h1>

```python

"""
    Convert wavs from data_origin/[machine] to color spectrograms into folders
    # construct dataset of images spectrogram (normal only) in folders:
        # data_png_cnn2/machines/[machine]
        # data_png_cnn2/all_png_test is used for images tests never seen by the model (1000 normal of each class & all anormal images)

"""
from os import listdir
from os.path import isfile, join
import numpy as np
import sys
import datetime
import shutil



# add parent folder to path 
from pathlib import Path
path = Path(os.getcwd())
str_parent = str(path.parent.absolute())
if not str_parent in sys.path:
    sys.path.append(str_parent)
import utils 
from SoundFile import SoundFile

```

```python
nbImagesTotestEachClass = 1000 # put this count of images in all_png_test_v5 (for each class)
start_time = datetime.datetime.now()
dictStat = {'normal': 0, 'anomaly': 0} # stats count of files in 2 classes

 
# create folders in order to export pngs files (remove all previous png) -> _data_png_cnn2/[machine]
mode = 0o777
cwd = os.getcwd() 
path = Path(cwd)

png_root_folder = str(path.parent.parent.absolute()) + '/_data_png_cnn2'
model_root_folder = str(path.parent.parent.absolute()) + '/_classifier_cnn2'
```

```python

# delete old folders if needed
try:
    shutil.rmtree(png_root_folder)
except OSError as e:
    print ("folder not found: %s - %s." % (e.filename, e.strerror))

try:
    shutil.rmtree(model_root_folder)
except OSError as e:
    print ("folder not found: %s - %s." % (e.filename, e.strerror))

os.mkdir(os.path.join(png_root_folder), mode)
print ("folder created: %s." % png_root_folder)
os.mkdir(os.path.join(model_root_folder), mode)
print ("folder created: %s." % model_root_folder)

os.mkdir(os.path.join(png_root_folder + '/all_png_test'), mode)
os.mkdir(os.path.join(png_root_folder + '/machines'), mode)

for folder_machine in utils.list_datasets: # ['valve']    
    os.mkdir(os.path.join(png_root_folder + '/machines/' + folder_machine), mode)
    
        
    
```

```python

out_folder_png_test = '../../_data_png_cnn2/all_png_test/'

for folder_machine in utils.list_datasets: # ['valve']
    for n in ['train', 'test']:    
        use_folder = '../../_data_origin/' + folder_machine + '/' + n + '/' 
        wavfiles = [f for f in listdir(use_folder) if isfile(join(use_folder, f))]
        if '.DS_Store' in wavfiles:
            wavfiles.remove('.DS_Store')
        
        nbWavs = len(wavfiles)
        countImages =  0    # counter for all images
        countImagesNormal =  0   
         
         # count files in each class
        for f in wavfiles:
            arrName = f.split("_") # anomaly_id_00_00000001.wav
            classPrefix = arrName[0] # 'normal' or 'anomaly'
            dictStat['normal'] =  dictStat['normal'] + 1 if classPrefix == 'normal' else dictStat['normal']

        print('dictStat: ', dictStat) # 479 anomaly 3291 normal

        # random choose of images for testing at the end: put them in png_test folder_machine
        arrIndicesImagesToTestNormal = np.random.randint(1, dictStat['normal'], nbImagesTotestEachClass) 

        for f in wavfiles:
            arrName = f.split("_") # anomaly_id_00_00000001.wav
            out_folder_png = ''
            classPrefix = arrName[0] # 'normal' or 'anomaly'
            out_folder_png = '../../_data_png_cnn2/machines/' + folder_machine + '/'
            
            countImages += 1
            if classPrefix == 'normal':
                countImagesNormal += 1

            # use some anomaly for the training set:
            if classPrefix == 'normal' and countImagesNormal in arrIndicesImagesToTestNormal or classPrefix == 'anomaly':
                out_folder_png = out_folder_png_test
            
            if countImages % 100 == 0:
                print('countImages generated...: ', countImages, ' / ', nbWavs, '(', folder_machine, ')')
        
            # print('use_folder: ', use_folder + f, out_folder_png, folder_machine)
            s = SoundFile(use_folder + f, out_folder_png, folder_machine)
            s.exportMelSpectrogramColor() # create color file    

now = datetime.datetime.now()
print("*************** End ******************", now.strftime("%Y-%m-%d %H:%M:%S"))
```
